```python
df_copy = df

df_copy.head()
```

```python
### Transforming my categorical columns

from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df_copy['Department'] = le.fit_transform(df_copy['Department'])
df_copy['salary'] = le.fit_transform(df_copy['salary'])
```

```python
df_copy.head()
```

```python
### Correlation Analysis

corr = df_copy.corr()

plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap = 'mako')
```

```python
### Splitting my data to x and y for my model

x = df_copy.drop(['left','number_project'],axis=1)
y = df_copy.left
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.2)
```

```python
x_test.head()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
### Test and Training my model

model.fit(x_train, y_train)
```

```python
model.predict(x_test)
```

```python
### Checking accuracy of my model

model.score(x_test, y_test)
```
